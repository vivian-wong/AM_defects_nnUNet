# Report for paper results
needs at least ~20 gigabytes of memory

```python
# recalculating mIOU cuz of the following issue: 
# in nnUNET, during preprocessing, the images were cropped to none zero regions. 
# i.e. the black parts outside our cylinder's rim are cropped off. 
# This isn't a problem in training (not sure)
# but it affects our accuracy in validation
# because in validation, they pad back the output so that those cut-off parts (pixel value = 0) are padded back
# but they pad it back to zero! 
# this is wrong because we want to pad it back to 1! since for us, 1 means black (i.e. true)!!!!!!!

# Therefore we need to crop the true mask to the size before padding. 
def get_indices_of_unpadded(y_pred, border_size = 100): 
    # y_pred: the 3D array of seg mask in nnUNet output; 0 = background/white. 1 = black
    # border_size: how many pixels should we look for padding? 
    corner_front_top_left = y_pred[:border_size,:border_size,:border_size]
    corner_back_bottom_right = y_pred[-border_size:,-border_size:,-border_size:]
    black_x, black_y, black_z = corner_front_top_left.nonzero()
    x1,y1,z1 = (black_x[0],black_y[0],black_z[0]) # indices where black pixels start
    black_x, black_y, black_z = (-border_size+arr for arr in corner_back_bottom_right.nonzero()) # in negative indices
    x2,y2,z2 = (black_x[-1], black_y[-1], black_z[-1]) # indices where the last black pixels are
    if x2 +1 != 0: 
        x2 = x2+1
    else: 
        x2 = None
    if y2 +1 != 0: 
        y2 = y2+1
    else: 
        y2 = None
    if z2 +1 != 0: 
        z2 = z2+1
    else: 
        z2 = None
    before_padding = y_pred[x1:x2,y1:y2,z1:z2]
#     assert before_padding.shape == (900,978,978) # i got this shape from log from preprocessing
    return x1,x2,y1,y2,z1,z2
```

```python
S1S2_pred_raw_paths_dict = {
#     '3D': "./results/nnUNet/3d_fullres/Task101_AM/nnUNetTrainerV2__nnUNetPlansv2.1/fold_0/validation_raw_postprocessed/S1S_2.nii.gz",
#     '2D': "./results/nnUNet/2d/Task101_AM/nnUNetTrainerV2__nnUNetPlansv2.1/fold_0/validation_raw_postprocessed/S1S_2.nii.gz",
    '3D Residual': "./results/nnUNet/3d_fullres/Task101_AM/nnUNetTrainerV2_ResencUNet__nnUNetPlans_FabiansResUNet_v2.1/fold_0/validation_raw_postprocessed/S1S_2.nii.gz"
}
```

```python
S1S2_true = nib.load("./dataset/nnUNet_raw_data/Task101_AM/labelsTr/S1S_2.nii.gz")
for model_name,pred_path in S1S2_pred_raw_paths_dict.items():
    S1S2_pred_raw = nib.load(pred_path)
    # turns out that raw and pp are exactly the same i.e. no postprocessing has been done.
    S1S2_true_data = S1S2_true.get_fdata().astype('uint8')
    S1S2_true.uncache()
    S1S2_pred_raw_data = S1S2_pred_raw.get_fdata().astype('uint8')
    S1S2_pred_raw.uncache()
    
    print(S1S2_true_data.shape)
    print(S1S2_pred_raw_data.shape)
    
    # undo padding
    x1,x2,y1,y2,z1,z2 = get_indices_of_unpadded(S1S2_pred_raw_data)
    S1S2_true_data_unpadded = S1S2_true_data[x1:x2,y1:y2,z1:z2]
    S1S2_pred_raw_data_unpadded = S1S2_pred_raw_data[x1:x2,y1:y2,z1:z2]
    
    print(S1S2_true_data_unpadded.shape, S1S2_true_data.dtype)
    print(S1S2_pred_raw_data_unpadded.shape, S1S2_pred_raw_data.dtype) # should both be less than (900, 978, 978) uint8
    
    # show example slice before and after crop, also shows prediction
    slice_number = 100
    fig, ax = plt.subplots(nrows=2,ncols=2, figsize = (40,40))
    ax[0,0].imshow(S1S2_true_data[slice_number,:,:],cmap='Greys')
    ax[0,0].title.set_text('Ground truth before cropping')
    ax[0,1].imshow(S1S2_true_data_unpadded[slice_number,:,:],cmap='Greys')
    ax[0,1].title.set_text('Ground truth after cropping')
    ax[1,0].imshow(S1S2_pred_raw_data[slice_number,:,:],cmap='Greys')
    ax[1,0].title.set_text('Prediction (before crop)')
    ax[1,1].imshow(S1S2_pred_raw_data_unpadded[slice_number,:,:],cmap='Greys')
    ax[1,1].title.set_text('Prediction (after crop)')
    fig.savefig('./{}_paper_figure_truth_prediction.png'.format(model_name))
    
    # delete padded images to free up memory
    del S1S2_pred_raw_data, S1S2_true_data
    S1S2_true_data_unpadded = S1S2_true_data_unpadded.flatten().astype('uint8')
    S1S2_pred_raw_data_unpadded = S1S2_pred_raw_data_unpadded.flatten().astype('uint8')
    
    print(model_name)
    # foreground IOU
    fIOU = jaccard_score(S1S2_true_data_unpadded, S1S2_pred_raw_data_unpadded)
    print('foreground IOU', fIOU)
#     background IOU 
    bIOU = jaccard_score(1-S1S2_true_data_unpadded, 1-S1S2_pred_raw_data_unpadded)
    print('background IOU', bIOU)
    print('mean IOU', (fIOU+bIOU)/2.0)
    print('--------------------------')
    

    
```
